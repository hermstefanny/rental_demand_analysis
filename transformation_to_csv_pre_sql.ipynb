```python
import pandas as pd
import numpy as np
```

```python
rental_data_part1 = pd.read_csv(r'housing_train.csv')
rental_data_part2 = pd.read_csv(r'housing_test.csv')

rental_data_flat = pd.concat([rental_data_part1, rental_data_part2])
print(len(rental_data_flat))
rental_data_flat = rental_data_flat.drop_duplicates()
rental_data_flat
```

```python
rental_data_flat.rename(columns = {'id':'ID',
                                   'url':'URL', 
                                   'region': 'Region',
                                   'region_url':'Region_URL',
                                   'rent': 'Rent',
                                   'type':'Type','sqfeet':'SqFeet','bedrooms':'Bedrooms','bathrooms':'Bathrooms','laundry_options':'Laundry_options', 'parking_options':'Parking_options', 'demand':'Demand','description':'Description','latitude':'Latitude','long':'Longitude', 'state':'State' }, inplace = True) 
```

```python
property_special_feat =rental_data_flat[['ID','cats_allowed', 'dogs_allowed', 'smoking_allowed',	'wheelchair_access',	'electric_vehicle_charge',	'comes_furnished']]
property_special_feat
```

```python
zero_instances = property_special_feat[(property_special_feat['cats_allowed'] == 0) & 
                                       (property_special_feat['dogs_allowed'] == 0) & 
                                       (property_special_feat['smoking_allowed'] == 0) & 
                                       (property_special_feat['wheelchair_access'] == 0) & 
                                       (property_special_feat['electric_vehicle_charge'] == 0) &
                                       (property_special_feat['comes_furnished'] == 0)]
zero_instances
```

```python
feat = list(property_special_feat.columns)
feat.remove('ID')
feat
features_id = range(1,len(feat)+1) 
for item in features_id:
    print(item)
```

```python
special_feat = pd.DataFrame({'FeatureID': features_id, 'Feature':feat})
print(special_feat)
```

```python
special_feat.to_csv('special_features.csv', index=False)
```

```python

for col in property_special_feat.columns[1:]:
    for index, row in special_feat.iterrows():
        if row['Feature'] == col:
            print(f'******* {col}, sucess!')
            property_special_feat.rename(columns = {col:row['FeatureID']}, inplace = True)
```

```python

for col in property_special_feat.columns[1:]:
    property_special_feat[col] = property_special_feat[col]*col

print(property_special_feat)
```

```python

df = property_special_feat.copy()
df.sort_values('ID')
```

```python
melted_df = pd.melt(df, id_vars='ID', var_name='Feature_ID', value_name='Value', ignore_index =False)
result_df = melted_df[melted_df['Value'] != 0][['ID', 'Feature_ID']]
result_df
```
